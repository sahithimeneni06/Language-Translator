```python
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer
import torch
```

```python
model_name = "facebook/mbart-large-50-many-to-many-mmt"

model = AutoModelForSeq2SeqLM.from_pretrained(
    model_name,
    torch_dtype=torch.bfloat16,
    device_map="auto"
)

tokenizer = AutoTokenizer.from_pretrained(
    model_name,
    src_lang="en_XX"
)

print(tokenizer.lang_code_to_id.keys())
```

```python
len(tokenizer.lang_code_to_id.keys())
```

```python
tokenizer.src_lang = "en_XX"
encoded = tokenizer("How are you?", return_tensors="pt")
generated_tokens = model.generate(
    **encoded,
    forced_bos_token_id=tokenizer.lang_code_to_id["te_IN"]
)

print(tokenizer.batch_decode(generated_tokens, skip_special_tokens=True))
```

```python
tokenizer.src_lang = "te_IN"

encoded = tokenizer("మీరు ఎలా ఉన్నారు?", return_tensors="pt").to(model.device)

generated_tokens = model.generate(
    **encoded,
    forced_bos_token_id=tokenizer.lang_code_to_id["en_XX"]
)

print(tokenizer.batch_decode(generated_tokens, skip_special_tokens=True))
```

```python

tokenizer.src_lang = "en_XX"

text = "Artificial Intelligence will change the world."

inputs = tokenizer(text, return_tensors="pt").to(model.device)

for lang_code in tokenizer.lang_code_to_id.keys():

    generated_tokens = model.generate(
        **inputs,
        forced_bos_token_id=tokenizer.lang_code_to_id[lang_code],
        max_length=60
    )

    translated_text = tokenizer.batch_decode(
        generated_tokens, skip_special_tokens=True
    )[0]

    print(f"{lang_code} → {translated_text}")
```
